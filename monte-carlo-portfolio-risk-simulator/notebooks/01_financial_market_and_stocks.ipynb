## 3. What is a Company?

A company is an organization created to produce goods or services with the objective of generating value.

As companies grow, they often require additional capital to:

- Develop new products
- Expand internationally
- Build factories
- Hire employees
- Invest in research and development

One way to obtain this capital is by selling ownership of the company.

### 3.1 From Company to Shareholders

When a private company decides to sell part of its ownership to investors, it can become a **public company**.

Instead of borrowing money from a bank, the company raises capital by issuing shares.

Each share represents a small fraction of ownership in the company.

The investors who purchase these shares become shareholders.

```text
                  Public Company
                        │
        ┌───────────────┴───────────────┐
        │                               │
     Issues Shares                  Raises Capital
        │
        ▼
  Investors Purchase Shares
        │
        ▼
 Become Company Shareholders
```

## 4. What is a Stock?

A stock represents partial ownership of a company.

For example, imagine a company divided into one million shares.

If you own one share, you own a very small percentage of that business.

Although your ownership may be small, you participate in the company's value creation.

As the company grows, investors may be willing to pay higher prices for its shares.

### Example

Suppose a company is divided into:

- 1,000,000 shares

If you own:

- 10,000 shares

then your ownership is:

$\frac{10,000}{1,000,000}=1\%$

This means you own **1% of the company**.

## 5. Why Do Stock Prices Change?

Unlike products in a store, stock prices constantly change.

Prices move because buyers and sellers continuously negotiate the value of a company.

Many factors influence these price movements, including:

- Company earnings
- Economic conditions
- Interest rates
- Investor sentiment
- Industry trends
- Unexpected news

As new information becomes available, investors adjust the price they are willing to pay.

```text
Company News
      │
      ▼
Investor Expectations
      │
      ▼
Buying and Selling Activity
      │
      ▼
Stock Price Changes
```

## 6. Understanding Market Data

For each stock, financial markets record:

- Opening price

- Lowest price
- Closing price
- Trading volume

These observations form a financial time series.

Historical market data is the foundation of quantitative finance and financial analytics.
