## 1. Business Context

Investing is a decision-making process under uncertainty.

Investors analyze historical information to understand market behavior, but historical performance does not guarantee future results.

The main challenge is not only estimating expected returns, but understanding the range of possible outcomes.

This project explores how financial analysts use Monte Carlo Simulation to model uncertainty and evaluate portfolio risk.

## 2. The Investment Problem

Imagine an investor with $100,000 USD who wants to invest in the stock market.

The investor wants to answer:

"How much could my portfolio be worth in one year?"

At first, this question seems simple.

However, financial markets are affected by multiple uncertain factors:

- Market movements.
- Economic conditions.
- Investor behavior.
- Company performance.
- Unexpected events.

## 3. Key Idea

**Traditional Forecast**

Today -> Future Value

**Monte Carlo Approach**

Today -> Scenario 1

Today -> Scenario 2

...

Today -> Scenario 10'000

## 4. Portfolio Scenario

Instead, capital is distributed across multiple companies to reduce concentration risk.
