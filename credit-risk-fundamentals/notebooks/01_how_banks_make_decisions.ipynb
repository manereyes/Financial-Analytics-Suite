## 1. Introduction

Every day, banks and financial institutions receive hundreds or even thousands of credit applications from customers requesting personal loans, auto financing, mortgages, and credit cards.


Approving every application would expose the institution to significant financial losses, while rejecting every application would eliminate valuable business opportunities.

To balance profitability and risk, banks evaluate the financial behavior of each applicant before making a lending decision.


This process is known as **Credit Risk Assessment**.


In this notebook, we will explore the business principles behind credit evaluation and build the intuition necessary to understand how modern credit scoring systems work.

## 4. What is Credit Risk?

Credit Risk is the possibility that a borrower fails to fulfill their financial obligations according to the agreed contract.


From the lender's perspective, every loan represents both:


- A business opportunity

- A potential financial loss

The objective of Credit Risk Management is not to eliminate risk completely, but to understand, quantify, and manage it.

## 5. What is a Credit Score?

A Credit Score is a numerical representation of a customer's creditworthiness.


Rather than measuring wealth, a credit score estimates the likelihood that a borrower will repay a financial obligation based on historical financial behavior.


Banks use credit scores to support decisions such as:


- Loan approvals
- Credit card approvals
- Credit limits
- Interest rates
- Portfolio segmentation
- Risk monitoring

## 7. Understanding Each Risk Factor

### 1. Payment History (35%)

Payment history is generally considered the most influential factor in credit evaluation.


Customers who consistently pay their financial obligations on time demonstrate financial discipline and lower credit risk.


Late payments, missed payments, or defaults significantly increase the perceived risk of future lending.

### 2. Credit Utilization (30%)

Credit utilization measures how much of the available credit is currently being used.


Lower utilization generally indicates healthier financial management.


Financial institutions commonly consider utilization below approximately 30% as a positive indicator.

### 3. Credit History Length (15%)

Managing different types of financial products demonstrates experience handling multiple financial obligations.


Examples include:


- Credit Cards
- Auto Loans
- Mortgages
- Personal Loans

### 5. Recent Credit Activity (10%)

Submitting numerous credit applications within a short period may indicate financial stress or increased borrowing needs.


Financial institutions often monitor recent credit inquiries as an additional risk indicator.
